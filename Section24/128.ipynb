## Handling Outliers

# 5 number Summary And Box Plot

```python
import numpy as np

lst_marks = [30, 40,55, 66, 78, 98, 89, 99, 100]
minm, Q1,median, Q3, maxm  = np.quantile(lst_marks, [0, 0.25, 0.50, 0.75, 1.0])
```

```python
minm, Q1, median, Q3, maxm
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_fence = Q1-(1.5*IQR)
higher_fence = Q3+(1.5*IQR)
lower_fence
```

```python
higher_fence
```

```python
import seaborn as sns
# Verticle boxplot
sns.boxplot(lst_marks)
```

```python
lst_marks = [-100,-200,200, 30, 40,55, 66, 78, 98, 89, 99, 100]
sns.boxplot(lst_marks)
```
