```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Create df

df = pd.DataFrame({
    'color' : ['red', 'green', 'blue', 'red', 'green', 'blue' ]
})
df.head()
```

```python
print(df[['color']].shape)  # 2d Panda.Dataframe
```

```python
encoder = OneHotEncoder()
# Perform fit and transform

type(encoder.fit_transform(df[['color']]).toarray()) # nd array
encoded = encoder.fit_transform(df[['color']]).toarray()  # also do alphabetical sorting
```

```python
import pandas as pd
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
encoded_df
```

```python
# For new data
encoder_df = encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df, encoded_df], axis=1)
```

```python
import pandas as pd
sns_df = pd.DataFrame({
    'color' : ['sex', 'smoker', 'day', 'time' ]
})
from sklearn.preprocessing import OneHotEncoder
sns_encoder = OneHotEncoder()
sns_encoded = sns_encoder.fit_transform(sns_df[['color']]).toarray()
```

```python
import pandas as pd
sns_encoded_df = pd.DataFrame(sns_encoded, columns=sns_encoder.get_feature_names_out())
sns_encoded_df
```
